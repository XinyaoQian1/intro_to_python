```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://dataprobing.com/col.jsp?id=119'
```

```python
header = {"user-agent":"Moilla/5.0"}
data = requests.get(url=url,headers=header)
```

```python
html=data.text
print(html)
```

```python
soup = BeautifulSoup(html,'html.parser')
lis = soup.find_all(name='div',attrs={"id":"row355"})
print(lis)
```

```python
i = 0
for li in lis:
    temp=li.find(name='div',attrs={'_index':str(i)})
    i=i+1
    print(temp.text)
```
